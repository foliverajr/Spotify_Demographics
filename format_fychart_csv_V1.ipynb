```python
import spotipy
import time
import numpy as np
import pandas as pd
from spotipy.oauth2 import SpotifyClientCredentials #To access authorised Spotify data

### Formats Spotify Chart CSV files to include audio features, popularity, and release date ###
def chart_expand(csv_path):
    client_id = 'e49e7cef01214924be4fb5e6fb232608'
    client_secret = '5bf23700fb5448c18daee8e6ef72db3e'
    client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)

    sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager) #spotify object to access API
    top_df = pd.read_csv(csv_path) #Makes the 2nd row the column headers, 1st row is junk
    acousticness = [] #Lists store specified attributes of tracks in the order they appear in top_df, so they will appear in the correct row
    danceability = []
    energy = []
    instrumentalness = []
    liveness = []
    loudness = []
    speechiness = []
    tempo = []
    valence = []
    popularity = []
    release_date = []
    songCount = 0
    totalRows = top_df.shape[0]
    for sp_id in top_df['spotify_id']:
        songCount += 1
        print(str(songCount) + '/' + str(totalRows))
        time.sleep(5)
        features = sp.audio_features(sp_id)
        track_info = sp.track(sp_id)
        acousticness.append(features[0]['acousticness'])
        danceability.append(features[0]['danceability'])
        energy.append(features[0]['energy'])
        instrumentalness.append(features[0]['instrumentalness'])
        liveness.append(features[0]['liveness'])
        loudness.append(features[0]['loudness'])
        speechiness.append(features[0]['speechiness'])
        tempo.append(features[0]['tempo'])
        valence.append(features[0]['valence'])
        popularity.append(track_info['popularity'])
        release_date.append(track_info['album']['release_date'])

    top_df['acousticness'] = acousticness #Creates the attribute columns and populates them
    top_df['danceability'] = danceability
    top_df['energy'] = energy
    top_df['instrumentalness'] = instrumentalness
    top_df['liveness'] = liveness
    top_df['loudness'] = loudness
    top_df['speechiness'] = speechiness
    top_df['tempo'] = tempo
    top_df['valence'] = valence
    top_df['popularity'] = popularity
    top_df['release_date'] = release_date
    top_df.to_csv(csv_path + '_formatted.csv')
    print('Done')
    
chart_expand('/home/fernando/top_200_annual.csv')
```
