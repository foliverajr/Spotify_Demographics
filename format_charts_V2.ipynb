```python
import spotipy
import numpy as np
import pandas as pd
import time
import multiprocessing as mp
from spotipy.oauth2 import SpotifyClientCredentials  # To access authorised Spotify data
from fycharts.SpotifyCharts import SpotifyCharts

num_parts = 20
num_cores = 8

def main(df, func):
    df_split = np.array_split(df, num_parts)
    pool = mp.Pool(mp.cpu_count())
    df = pd.concat(pool.map(func, df_split))
    pool.close()
    pool.join()
    return df

### Formats Spotify Chart CSV files to include audio features, popularity, and release date ###


def chart_expand(top):
    songc=1
    client_id = 'e49e7cef01214924be4fb5e6fb232608'
    client_secret = '5bf23700fb5448c18daee8e6ef72db3e'
    client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)

    sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)  # spotify object to access API
    acousticness = []  # Lists store specified attributes of tracks in the order they appear in top_df, so they will appear in the correct row
    danceability = []
    energy = []
    instrumentalness = []
    liveness = []
    loudness = []
    speechiness = []
    tempo = []
    valence = []
    popularity = []
    release_date = []
    featured_artists=[]

    for sp_id in top['spotify_id']:
        ft=""
        main_artists=[]
        time.sleep(3)
        print(songc, end='|')
        songc+=1
        features = sp.audio_features(sp_id)
        track_info = sp.track(sp_id)
        acousticness.append(features[0]['acousticness'])
        danceability.append(features[0]['danceability'])
        energy.append(features[0]['energy'])
        instrumentalness.append(features[0]['instrumentalness'])
        liveness.append(features[0]['liveness'])
        loudness.append(features[0]['loudness'])
        speechiness.append(features[0]['speechiness'])
        tempo.append(features[0]['tempo'])
        valence.append(features[0]['valence'])
        popularity.append(track_info['popularity'])
        release_date.append(track_info['album']['release_date'])
        for item in range(len(track_info['artists'])):
            if item == 0 and track_info['album']['artists'][0]['name'] == 'Various Artists':
                continue
            if track_info['artists'][item]['name'] not in track_info['album']['artists'][0]['name']:
                ft=ft+track_info['artists'][item]['name']+', '
        featured_artists.append(ft)
    featured_artists[-1][:-2] # Removes the last ',' after the last featured artist
    top['featured-artist'] = featured_artists
    top['acousticness'] = acousticness  # Creates the attribute columns and populates them
    top['danceability'] = danceability
    top['energy'] = energy
    top['instrumentalness'] = instrumentalness
    top['liveness'] = liveness
    top['loudness'] = loudness
    top['speechiness'] = speechiness
    top['tempo'] = tempo
    top['valence'] = valence
    top['popularity'] = popularity
    top['release_date'] = release_date
    # top.to_csv('200Annual_formatted.csv')
    print('Done')
    return top



if __name__ == "__main__":
    SpotifyCharts().viral50Daily(output_file="200_90.csv", start="2019-01-01", end="2019-01-01", region=['global'])
    time.sleep(3)
    top_df = pd.read_csv('200_90.csv')
    print(top_df)
    start_time = time.time()
    qwe = main(top_df, chart_expand)
    print('DONE')
    print("--- %s seconds ---" % (time.time() - start_time))
    qwe.to_csv('200Annual_formatted.csv')
```
